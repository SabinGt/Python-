## Seaborn Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
sns.set()
```

```python
new_iris=sns.load_dataset('iris')
```

```python
new_iris.head(5)
```

```python
new_iris.petal_width.plot()
```

```python
new_iris.plot()
```

```python
new_d = pd.Series(np.random.rand(7), index = list('ABCDEFG'))
```

```python
new_d
```

```python
new_d.plot.bar(color='r',alpha=1)
```

```python
new_d.plot.barh(color='b',alpha=0.5)
```

```python
new_load = sns.load_dataset('tips')
```

```python
new_load.head()
```

```python
sizeOfDay = pd.crosstab(new_load["day"],new_load["size"])
```

```python
sizeOfDay
```

```python
sizeOfDay.plot.bar()
```

```python
sizeOfDay.plot.hist()
```
